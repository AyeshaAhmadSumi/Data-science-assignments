```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
magnitudes = [1.20, 0.40, 0.76, 0.78, 0.87, 1.12, 1.30, 1.00, 0.90, 1.40]
```

```python
# Find the maximum magnitude
max_magnitude = max(magnitudes)
```

```python
# Calculate and print the percentage for each magnitude
for mag in magnitudes:
    percentage = (mag / max_magnitude) * 100
    print(f"Magnitude: {mag} ‚Üí {percentage:.2f}% of max magnitude")
```

```python
magnitudes = [1.2, 0.4, 0.76, 0.78, 0.87, 1.12, 1.3, 1.0, 0.9, 1.4]
percentages = [85.71, 28.57, 54.29, 55.71, 62.14, 80.00, 92.86, 71.43, 64.29, 100.00]
locations = [
    "Big Spring, Texas", "The Geysers, CA", "The Geysers, CA", "The Geysers, CA", "Anza, CA",
    "Corona, CA", "Beluga, Alaska", "Cobb, CA", "Clear, Alaska", "Funny River, Alaska"
]

```

```python
# bar graph
plt.figure(figsize=(10, 6))
plt.barh(locations, percentages, color='skyblue')

# Add titles and labels
plt.title("Earthquake magnitude in different locations", fontsize=14)
plt.xlabel("Percentage (%)", fontsize=12)
plt.ylabel("Location", fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
#pie chart
plt.figure(figsize=(8, 8))
plt.pie(percentages, labels=locations, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)

# Add title
plt.title("Earthquake magnitude in different locations", fontsize=14)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# box plot
ax.boxplot(percentages, vert=False, patch_artist=True, boxprops=dict(facecolor='skyblue', color='blue'),
           medianprops=dict(color='red'))

# Set title and labels
ax.set_title("Earthquake magnitude in different locations", fontsize=14)
ax.set_xlabel("Percentage (%)", fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting the line chart
ax.plot(locations, percentages, marker='o', linestyle='--' , color='green')

# Adding title and labels
ax.set_title("Earthquake magnitude in different locations", fontsize=14)
ax.set_xlabel("Location", fontsize=12)
ax.set_ylabel("Percentage (%)", fontsize=12)

# Rotating x-axis labels for better readability
plt.xticks(rotation=360)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting the scatter plot
ax.scatter(locations, percentages, color='magenta', edgecolor='black', s=100)

# Adding title and labels
ax.set_title("Earthquake magnitude in different locations", fontsize=14)
ax.set_xlabel("Location", fontsize=10)
ax.set_ylabel("Percentage (%)", fontsize=12)

# Rotating x-axis labels for better readability
plt.xticks(rotation=360)

# Display the plot
plt.tight_layout()
plt.show()
```
